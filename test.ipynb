```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt

q,p,L = sp.symbols('q p L')

q = sp.Eq(4.8 - 0.5*p,q) 

p = sp.solve(q,p)

display(p)

q = np.linspace(0,10,100)

def p(q):
    return 9.2 - 2*q

fig, ax = plt.subplots()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(q,p(q))
ax.axhline(0);
```
